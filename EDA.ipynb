## Exploratory data analysis

```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns; sns.set(style="ticks", color_codes=True)
from sklearn import datasets
iris = datasets.load_iris()
irisDF = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['species'])
```

```python
# Quick View
irisDF.head(5)
```

```python
# Checking the types
irisDF.dtypes
```

The class column in this dataset contains the values 0.0, 1.0, 2.0 but "species" are categories corresponding to the classes associated with each row (Setosa, Versicolor and Virginica)

```python
# Describing the attributes (in this case all are numeric)
irisDF.describe()
```

```python
# Plot histograms to discover distributions... one particular variable.
irisDF['sepal length (cm)'].plot.hist()
```

```python
# Plotting all histograms.
irisDF.hist()
```

```python
# Seaborn pairplots are usefull to look pairwise relationships between the attributes 
# The principal diagonal shows the histograms of each variable
g = sns.pairplot(irisDF, hue="species", markers=["o", "s", "D"])
```

```python
# Visualizing each variable boxplot
irisDF.boxplot()
```

```python
# Check the boxplot by class 
irisDF.boxplot(by = "species")
```

```python
# Seaborn points with boxes 
sns.boxplot(x="species", y="petal length (cm)", data=irisDF)
sns.swarmplot(x="species", y="petal length (cm)", data=irisDF, color = ".55")
```
